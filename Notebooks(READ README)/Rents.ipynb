## 1. Organize the data : Commercial Real Estate Price 2014-2019

```python
#Get all five data sets
df1 = pd.read_csv('Price_1.csv',encoding='cp949') #2014-2015
print(df1.shape)
df1.head()
```

```python
df2 = pd.read_csv('Price_2.csv',encoding='cp949') #2015-2016
print(df2.shape)
df2.head()
```

```python
df3 = pd.read_csv('Price_3.csv',encoding='cp949') #2016-2017
print(df3.shape)
df3.head()
```

```python
df4 = pd.read_csv('Price_4.csv',encoding='cp949') #2017-2018
print(df4.shape)
df4.head()
```

```python
df5 = pd.read_csv('Price_5.csv',encoding='cp949') #2018-2019
print(df5.shape)
df5.head()
```

```python
#Define a function that cleans theses datasets
def clean_df(x=pd.DataFrame):
    x = x[['시군구','전용/연면적(㎡)','거래금액(만원)']] # Get only necessary features
    x.columns = ['Address','Space','Price'] #Translate
    x['Price']=x['Price'].str.replace(',','') #Remove commas and convert in to float value
    x['Price']=x['Price'].astype(float) 
    x= x.groupby(['Address']).sum() #Group by address to reduce rows and cut down runtime for append
    x.reset_index(inplace=True)
    return x
```

```python
#Clean all five datasets
df1=clean_df(df1)
df2=clean_df(df2)
df3=clean_df(df3)
df4=clean_df(df4)
df5=clean_df(df5)
```

```python
df1=df1.append([df2,df3,df4,df5],ignore_index=True) #Merge all the dataframes
print(df1.shape)
df1.head()
```

```python
df_price= df1.groupby(['Address']).sum() #Group by address and get the average values
df_price.reset_index(inplace=True)
df_price['Fee_Space']=df_price['Price']/df_price['Space']
df_price.drop(columns=['Space','Price'],inplace=True)
print(df_price.shape)
df_price.head()
```
